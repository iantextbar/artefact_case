```python
import datetime
import sys
sys.path.append('../../')
from src.agents.agent import agent
```

```python
def conversation(phone, date, day_of_week):

    config = {
    "configurable": {
        "thread_id": "whatsapp-session-001",
        "client_phone": phone
        }
    }

    print("🎸 [Slash]: Fala, rocker! Digite 'sair' para encerrar.\n")

    while True:
        user_input = input("Você: ")
        print(user_input)
        if user_input.strip().lower() == 'sair':
            print("\n🎸 [Slash]: Valeu pelo papo! Let's go rockers!! 🤘")
            break

        if not user_input.strip():
            continue

        metadata_context = (
            f"{{Telefone Cliente: {phone}, "
            f"Dia da Semana: {day_of_week}, "
            f"Horário: {date}}}"
        )

        full_prompt_input = f"Metadados: {metadata_context}\nCliente: {user_input}"

        try:
            response = agent.invoke(
                {
                    "messages": [{"role": "user", "content": full_prompt_input}]
                },
                config=config
            )

            raw_content = response["messages"][-1].content

            raw_content = response["messages"][-1].content
            if isinstance(raw_content, list):
                final_reply = "".join([block.get("text", "") for block in raw_content if block.get("type") == "text"])
            else:
                final_reply = raw_content

            print(f"\n🎸 Slash: {final_reply}\n")
            print("-" * 80)

        except Exception as e:
            print(f"\n❌ Microfonia na jam (Erro): {str(e)}\n")
```

```python
CLIENT_PHONE = "(67) 99890-1234"

agora = datetime.datetime.now()
dias_semana = ["Segunda-feira", "Terça-feira", "Quarta-feira", "Quinta-feira", "Sexta-feira", "Sábado", "Domingo"]
time = agora.strftime('%H:%M')
day_of_week = dias_semana[agora.weekday()]

conversation(CLIENT_PHONE, time, day_of_week)
```
